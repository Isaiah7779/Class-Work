```python
from sklearn.datasets import fetch_california_housing
import matplotlib.pyplot as plt
from matplotlib import style
%matplotlib inline
```

```python
c_h = fetch_california_housing()
```

```python
print(c_h)
```

```python
x_axis = c_h.data
```

```python
y_axis = c_h.target
```

```python
style.use('ggplot')
plt.figure(figsize=(7,7))
plt.hist(y_axis, bins=50)
plt.xlabel("Price")
plt.ylabel("Numbers")
plt.show
```

```python
style.use('ggplot')
plt.figure(figsize=(7,7))
plt.scatter(c_h.data[:,1],c_h.target)
plt.xlabel("Price")
plt.ylabel("Numbers")
plt.show
```
