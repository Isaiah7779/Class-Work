```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
cie = np.array([85, 90, 75, 80, 95, 70, 65, 78, 88, 92])
see = np.array([88, 92, 78, 85, 96, 75, 70, 82, 90, 94])
```

```python
X_train, X_test, y_train, y_test = train_test_split(cie.reshape(-1, 1), see, test_size=0.2, random_state=42)
```

```python
model = LinearRegression()
```

```python
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
```

```python
print("Linear Regression Coefficients:", model.coef_)
print("Mean Squared Error:", mse)
print("R-squared:", r2)
```

```python
plt.scatter(X_test, y_test, color='blue')
plt.plot(X_test, y_pred, color='red', linewidth=2)
plt.xlabel('CIE Scores')
plt.ylabel('SEE Scores')
plt.title('Linear Regression: CIE vs SEE')
plt.show()
```
