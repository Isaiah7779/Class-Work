```python
import numpy as np
import matplotlib.pyplot as plt

dataset = [11, 10, 12, 12, 34, 23, 21, 12, 18, 18, 34, 54, 102, 89, 64, 85, 36, 65, 78, 88, 23, 11, 23, 12, 23, 24, 25, 65, 23, 45, 89, 75, 54, 56, 35, 88, 67, 69, 45]

def detect_outliers(data):
    outliers = []
    threshold = 1.5  # Changed the threshold to 1.5 (a common value for outliers)
    q75, q25 = np.percentile(data, [75 ,25])
    iqr = q75 - q25

    for i in data:
        if i < (q25 - 1.5 * iqr) or i > (q75 + 1.5 * iqr):
            outliers.append(i)

    return outliers

outlier_pts = detect_outliers(dataset)
print("Outliers:", outlier_pts)
```

```python

plt.boxplot(dataset)
plt.title('Box plot of the dataset')
plt.ylabel('Values')
plt.show()
```
